```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
timesnet_dis = pd.read_csv("results/TimesNet/disjoint.csv")
timesnet_avr = pd.read_csv("results/TimesNet/ensemble_32_20_overlapping.csv")

dlinear_dis = pd.read_csv("results/DLinear/ disjoint_short.csv")
dlinear_avr = pd.read_csv("results/DLinear/ mean.csv")

linear_dis = pd.read_csv("results/Linear/disjoint_3.csv")
linear_avr = pd.read_csv("results/Linear/overlapping_all_3.csv")


itrans_dis = pd.read_csv("results/iTransformer/disjoint.csv")
itrans_avr = pd.read_csv("results/iTransformer/mean.csv")

trans_dis = pd.read_csv("results/Transformer/disjoint.csv")
trans_avr = pd.read_csv("results/Transformer/mean.csv")

# ---- AUC-PR summary for each model & strategy ----
summary = pd.DataFrame({
    "Model":    ["TimesNet"]*2 + ["DLinear"]*2 + ["Linear"]*2 + ["iTransformer"]*2  + ["Transformer"]*2,
    "Strategy": ["disjoint", "mean"] * 5,
    "AUC-PR": [
        timesnet_dis["AUC-PR"].mean(),
        timesnet_avr["AUC-PR"].mean(),
        dlinear_dis["AUC-PR"].mean(),
        dlinear_avr["AUC-PR"].mean(),
        linear_dis["AUC-PR"].mean(),
        linear_avr["AUC-PR"].mean(),
        itrans_dis["AUC-PR"].mean(),
        itrans_avr["AUC-PR"].mean(),
        trans_dis["AUC-PR"].mean(),
        trans_avr["AUC-PR"].mean()
    ]
})

# nicely formatted table (model x strategy)
print(summary.pivot(index="Model", columns="Strategy", values="AUC-PR").round(2))
```

```python
# --- Compute means ---
strategies = ["Disjoint", "Overlapping"]

results = {
    "TimesNet": [
        timesnet_dis["AUC-PR"].mean(),
        timesnet_avr["AUC-PR"].mean()
    ],
    "DLinear": [
        dlinear_dis["AUC-PR"].mean(),
        dlinear_avr["AUC-PR"].mean()
    ],
    "Linear": [
        linear_dis["AUC-PR"].mean(),
        linear_avr["AUC-PR"].mean()
    ],
    "iTransformer": [
        itrans_dis["AUC-PR"].mean(),
        itrans_avr["AUC-PR"].mean()
    ],
    "Transformer": [
        trans_dis["AUC-PR"].mean(),
        trans_avr["AUC-PR"].mean()
    ]
}

# --- Plot ---
plt.figure(figsize=(9,5))

for model, scores in results.items():
    plt.scatter(strategies, scores, s=90, alpha=0.9)      # scatter points
    plt.plot(strategies, scores, label=model, alpha=0.7)  # line to highlight evolution

plt.hlines(0.37, strategies[0], strategies[-1], ls="dashed", color="gray", label="Sub-PCA")
plt.title("AUC-PR Evolution Across Strategies")
plt.xlabel("Strategy")
plt.ylabel("AUC-PR")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig("strategies.pdf", format="pdf", bbox_inches='tight')
plt.show()
```

```python
names = dlinear_avr["filename"]
ucr_idx = [i for i, name in enumerate(names) if "UCR" in name]
timesnet_avr.loc[ucr_idx].mean(numeric_only=True)


models = {
    "DLinear": dlinear_avr,
    "TimesNet": timesnet_avr,
    "iTransformer": itrans_avr,
    "Linear": linear_avr,
    "Transformer": trans_avr,
}



first_model = next(iter(models))
filenames = names[ucr_idx]

# Build matrix of AUC-PR values
data = []
model_names = []

for model_name, df in models.items():
    # align on filename, just in case order differs
    s = (
        df.set_index("filename")["Standard-F1"]
          .reindex(filenames)          # ensures same order
    )
    data.append(s.values)
    model_names.append(model_name)

data = np.array(data)   # shape: (n_models, n_files)
n_models, n_files = data.shape


# ------------------------------
# 2. Define Plotting Parameters
# ------------------------------
center_radius = 0.25      # radius of the empty hole
inner_heatmap_radius = 0.45  # where the first ring starts
outer_heatmap_radius = 1.0    # outer boundary

# radii for rings
r = np.linspace(inner_heatmap_radius, outer_heatmap_radius, n_models + 1)
theta = np.linspace(0, 2*np.pi, n_files + 1)

R, Theta = np.meshgrid(r, theta)

# ------------------------------
# 3. Define Discrete Colormap and Norm
# ------------------------------
N_LEVELS = 10
V_MIN, V_MAX = 0.0, 1.0
base_cmap = plt.cm.get_cmap('viridis')
cmap_colors = base_cmap(np.linspace(0, 1, N_LEVELS))
cmap = colors.ListedColormap(cmap_colors)

bounds = np.linspace(V_MIN, V_MAX, N_LEVELS + 1)
norm = colors.BoundaryNorm(bounds, cmap.N)

cbar_ticks = bounds[:-1] + (bounds[1] - bounds[0])# / 2

# ------------------------------
# 4. Plot
# ------------------------------
plt.rcParams.update({'font.size': 14})
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="polar")
ax.grid(False)
pc = ax.pcolormesh(
    Theta, R, data.T,
    shading="auto",
    cmap=cmap,
    norm=norm,
    edgecolor='grey', # ADDED: Grey edges
    linewidth=0.5,    # ADDED: Edge linewidth
    rasterized=True
)

# ------------------------------
# 5. Colorbar
# ------------------------------
cbar = plt.colorbar(
    pc,
    ax=ax,
    pad=0.1,
    orientation="vertical",
    ticks=cbar_ticks,
    boundaries=bounds,
    shrink=0.5
)

cbar.set_ticklabels([f'{t:.2f}' for t in cbar_ticks])
cbar.set_label("AUC-PR", fontsize=16, weight='bold')

# ------------------------------
# 6. Aesthetic Touches
# ------------------------------
ax.set_theta_direction(-1)
ax.set_theta_offset(np.pi/2)
ax.set_xticklabels([])
ax.set_yticklabels([])

ax.set_ylim(0, outer_heatmap_radius)


# Add Model Names (Radial Labels)
ring_centers = (r[:-1] + r[1:]) / 2
label_pos_r = ring_centers
label_angle = np.pi #np.pi/2 + 0.15 # Offset slightly from the top vertical line

for i, name in enumerate(model_names):
    ax.text(label_angle, label_pos_r[i], name,
            ha='center', va='center',
            bbox=dict(facecolor='white', alpha=0.5, edgecolor='none', boxstyle='round,pad=0.05'),
            zorder=12,
            fontsize=12,
            rotation=0#-90
           )


plt.tight_layout()
plt.savefig("polar_heatmap.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

dataset_names = ["UCR", "YAHOO", "WSD"]#, "Exathlon", "SMD", "NAB", "Stock", "IOPS"]

models = {
    "Linear": linear_avr,
    "TimesNet": timesnet_avr,
    "iTransformer": itrans_avr,
    "Transformer": trans_avr,
}

plt.figure(figsize=(12, 6))

n_datasets = len(dataset_names)
n_models = len(models)

# How far to shift each model's violin from the dataset center
offset = 0.15  

for m_idx, (model_name, df) in enumerate(models.items()):
    data = []
    positions = []

    for d_idx, ds in enumerate(dataset_names):
        # Select rows for this dataset
        mask = df["filename"].str.contains(ds)
        vals = df.loc[mask, "AUC-ROC"].dropna()
        print(model_name, ds, np.mean(vals))

        if len(vals) > 0:
            data.append(vals)
            # Center of dataset is (d_idx + 1); we shift each model left/right
            pos = d_idx + 1 + (m_idx - (n_models - 1) / 2) * offset
            positions.append(pos)

    if len(data) == 0:
        continue

    parts = plt.violinplot(
        data,
        positions=positions,
        widths=offset * 0.9,
        showmeans=True,
        showmedians=False,
        showextrema=False,
    )

    # Make violins a bit transparent so overlaps are visible
    for pc in parts["bodies"]:
        pc.set_alpha(0.4)

    # Store one patch per model for legend (color taken from first body)
    color = parts["bodies"][0].get_facecolor().flatten()
    if m_idx == 0:
        legend_patches = []
    legend_patches.append(mpatches.Patch(color=color, label=model_name))

# X-axis: one tick per dataset
plt.xticks(range(1, n_datasets + 1), dataset_names)
plt.ylabel("AUC-PR")
plt.title("Performance Comparison Across Models and Datasets")

plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend(handles=legend_patches, title="Model")
plt.tight_layout()
plt.savefig("datasets.pdf", format="pdf")
plt.show()
```
