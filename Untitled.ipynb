```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
timesnet_mse = pd.read_csv("results/TimesNet/MSE.csv")
timesnet_dis = pd.read_csv("results/TimesNet/disjoint.csv")
timesnet_avr = pd.read_csv("results/TimesNet/mean.csv")

dlinear_mse = pd.read_csv("results/DLinear/MSE.csv")
dlinear_dis = pd.read_csv("results/DLinear/ disjoint_short.csv")
dlinear_avr = pd.read_csv("results/DLinear/ mean.csv")

linear_mse = pd.read_csv("results/Linear/MSE_3.csv")
linear_dis = pd.read_csv("results/Linear/disjoint_3.csv")
linear_avr = pd.read_csv("results/Linear/overlapping_all_3.csv")


itrans_mse = pd.read_csv("results/iTransformer/MSE.csv")
itrans_dis = pd.read_csv("results/iTransformer/disjoint.csv")
itrans_avr = pd.read_csv("results/iTransformer/mean.csv")

trans_mse = pd.read_csv("results/Transformer/MSE.csv")
trans_dis = pd.read_csv("results/Transformer/disjoint.csv")
trans_avr = pd.read_csv("results/Transformer/mean.csv")

# ---- AUC-PR summary for each model & strategy ----
summary = pd.DataFrame({
    "Model":    ["TimesNet"]*3 + ["DLinear"]*3 + ["Linear"]*3 + ["iTransformer"]*3  + ["Transformer"]*3,
    "Strategy": ["MSE", "disjoint", "mean"] * 5,
    "AUC-PR": [
        timesnet_mse["AUC-PR"].mean(),
        timesnet_dis["AUC-PR"].mean(),
        timesnet_avr["AUC-PR"].mean(),
        dlinear_mse["AUC-PR"].mean(),
        dlinear_dis["AUC-PR"].mean(),
        dlinear_avr["AUC-PR"].mean(),
        linear_mse["AUC-PR"].mean(),
        linear_dis["AUC-PR"].mean(),
        linear_avr["AUC-PR"].mean(),
        itrans_mse["AUC-PR"].mean(),
        itrans_dis["AUC-PR"].mean(),
        itrans_avr["AUC-PR"].mean(),
        trans_mse["AUC-PR"].mean(),
        trans_dis["AUC-PR"].mean(),
        trans_avr["AUC-PR"].mean()
    ]
})

# nicely formatted table (model x strategy)
print(summary.pivot(index="Model", columns="Strategy", values="AUC-PR").round(2))
```

```python
# --- Compute means ---
strategies = ["MSE", "Disjoint", "Overlapping"]

results = {
    "TimesNet": [
        timesnet_mse["AUC-PR"].mean(),
        timesnet_dis["AUC-PR"].mean(),
        timesnet_avr["AUC-PR"].mean()
    ],
    "DLinear": [
        dlinear_mse["AUC-PR"].mean(),
        dlinear_dis["AUC-PR"].mean(),
        dlinear_avr["AUC-PR"].mean()
    ],
    "Linear": [
        linear_mse["AUC-PR"].mean(),
        linear_dis["AUC-PR"].mean(),
        linear_avr["AUC-PR"].mean()
    ],
    "iTransformer": [
        itrans_mse["AUC-PR"].mean(),
        itrans_dis["AUC-PR"].mean(),
        itrans_avr["AUC-PR"].mean()
    ],
    "Transformer": [
        trans_mse["AUC-PR"].mean(),
        trans_dis["AUC-PR"].mean(),
        trans_avr["AUC-PR"].mean()
    ]
}

# --- Plot ---
plt.figure(figsize=(9,5))

for model, scores in results.items():
    plt.scatter(strategies, scores, s=90, alpha=0.9)      # scatter points
    plt.plot(strategies, scores, label=model, alpha=0.7)  # line to highlight evolution

plt.hlines(0.37, strategies[0], strategies[-1], ls="dashed", color="gray", label="Sub-PCA")
plt.title("AUC-PR Evolution Across Strategies")
plt.xlabel("Strategy")
plt.ylabel("AUC-PR")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig("strategies.pdf", format="pdf", bbox_inches='tight')
plt.show()
```
