```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_timesnet_ensemble = pd.read_csv("results/TimesNet/ensemble_32_20_overlapping_noschedule_test.csv")
df_timesnet_ensemble.mean(numeric_only=True).round(3) * 100
```

```python
df_timesnet_ensemble_ml = pd.read_csv("results/TimesNet/ensemble_ML_128_20_overlapping.csv")
df_timesnet_ensemble_ml.mean(numeric_only=True).round(3) * 100
```

```python

for metr in ["AUC-PR", "AUC-ROC", "Standard-F1"]:
    over_rows = []
    for seed in range(5):
        over = pd.read_csv(f"results/TimesNet/32_20_overlapping_{seed}_strides.csv")
        over = over[over["stride"] == 1]
        over_rows.append(over[metr])
    
    over_arr = np.stack(over_rows, axis=0)
    over_row_mean = over_arr.mean(axis=0)
    over_row_std  = over_arr.std(axis=0)
    
    over_global_mean = over_row_mean.mean()
    over_global_std  = over_row_std.mean()

    print(f"{metr}: {over_global_mean*100:.1f} ± {over_global_std*100:.1f}")
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# --- global "print-quality" settings ---
mpl.rcParams.update({
    "figure.dpi": 140,          # notebook display
    "savefig.dpi": 300,         # export quality
    "font.size": 11,
    "axes.titlesize": 13,
    "axes.labelsize": 11,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "legend.fontsize": 10,
    "axes.linewidth": 1.0,
    "lines.linewidth": 2.0,
})

# --- data ---
d = df_timesnet_ensemble[["rel_abnormal_mean", "rel_normal_mean", "AUC-PR"]].dropna()
x = (d["rel_abnormal_mean"] / d["rel_normal_mean"]).to_numpy()
y = d["AUC-PR"].to_numpy()
mask = np.isfinite(x) & np.isfinite(y) & (x > 0)
x, y = x[mask], y[mask]
lx = np.log10(x)

# correlations
pearson_log = np.corrcoef(lx, y)[0, 1]
rx = lx.argsort().argsort().astype(float)
ry = y.argsort().argsort().astype(float)
spearman = np.corrcoef(rx, ry)[0, 1]

# trend line (y ~ a log10(x) + b)
a, b = np.polyfit(lx, y, 1)
grid = np.linspace(lx.min(), lx.max(), 300)
x_grid = 10**grid
y_hat = a * grid + b

# --- plot ---
fig, ax = plt.subplots(figsize=(7.5, 4.5), constrained_layout=True)

ax.scatter(x, y, s=10, alpha=0.25, rasterized=True)  # rasterized keeps PDF small
ax.plot(x_grid, y_hat)

ax.axvline(1.0, linestyle="--", linewidth=1)

ax.set_xscale("log")
ax.set_xlabel("Ratio: rel_abnormal_mean / rel_normal_mean")
ax.set_ylabel("AUC-PR")
ax.set_title("AUC-PR vs relative-error ratio")

ax.grid(True, which="major", axis="x", alpha=0.3)
ax.grid(True, which="major", axis="y", alpha=0.15)

ax.text(
    0.02, 0.98,
    f"Pearson r (log10 ratio) = {pearson_log:.3f}\nSpearman ρ = {spearman:.3f}",
    transform=ax.transAxes,
    va="top",
    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="0.7")
)
plt.ylim(-0.1, 1.1)
plt.xlim(0.05, 10000)
# --- export (pick one) ---
fig.savefig("aucpr_vs_ratio.svg")   # best for print (vector)
fig.savefig("aucpr_vs_ratio.pdf")   # best for LaTeX/papers
fig.savefig("aucpr_vs_ratio.png")   # good for slides
plt.show()
```
