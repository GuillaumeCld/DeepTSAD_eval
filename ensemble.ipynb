```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
def filter(df, files):
    return df[df["filename"].isin(files)]
```

```python
file_list_path = 'Datasets/File_List/TSB-AD-U-Eva.csv'
file_list = pd.read_csv(file_list_path)['file_name'].values
```

```python
df = pd.read_csv("results/DLinear/ensemble_32_20_disjoint.csv")
filter(df, file_list).mean(numeric_only=True).round(3)
```

```python
df = pd.read_csv("results/DLinear/ensemble_32_20_overlapping.csv")
filter(df, file_list).mean(numeric_only=True).round(3)
```
